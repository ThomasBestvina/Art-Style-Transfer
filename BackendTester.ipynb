```python
import matplotlib.pyplot as plt
import requests
import base64
import json  
import os
import shutil
import io
import urllib
import numpy as np
import PIL
```

```python
url = 'http://localhost:5000/processing'
path_img = '/home/thomas/Pictures/99826-1602532255.jpg'

gennum = 0

url += "/"+ str(gennum)

with open(path_img, 'rb') as img:
  name_img= os.path.basename(path_img)
  files= {'image': (name_img,img,'multipart/form-data',{'Expires': '0'}) }
  with requests.Session() as s:
    r = s.post(url,files=files)
    image = np.array(PIL.Image.open(urllib.request.urlopen(r.text[10:-2])))
    plt.imshow(image)
    plt.show()
```
